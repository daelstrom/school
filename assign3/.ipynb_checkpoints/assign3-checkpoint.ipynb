### 3. Pre-process the data
     0   Sl_No                660 non-null    int64    [index]
     1   Customer Key         660 non-null    int64    (removed)
     2   Avg_Credit_Limit     660 non-null    int64
     3   Total_Credit_Cards   660 non-null    int64
     4   Total_visits_bank    660 non-null    int64
     5   Total_visits_online  660 non-null    int64
     6   Total_calls_made     660 non-null    int64
 
 No need to keep the Customer Key as they are merely unique identifiers whereas the serial identification number implies a 

```python
# setting index, serial number should not be included
# within the actual dataset
customer_data_train.set_index("Sl_No", inplace=True)

# scaling, first to normalize distances between points
# (standard scaling)
num_pipeline = Pipeline([
  ("scaler", StandardScaler())  
])

# then adding t
full_pipeline = ColumnTransformer([
    ( "num", num_pipeline, 
        [
            "Customer Key", 
            "Avg_Credit_Limit", 
            "Total_Credit_Cards", 
            "Total_visits_bank", 
            "Total_visits_online", 
            "Total_calls_made"
        ]
    )
])

# fitting pipeline to training data
full_pipeline.fit(customer_data_train)

# transforming training and testing data
customer_train_x = full_pipeline.fit_transform(customer_data_train)
```

### 4. Train the model

```python
# k-means model
# evaluating hyperparameter (k) values 
k = 5 
kmeans = KMeans(n_clusters=k, n_init=100)
customer_predict_y = kmeans.fit_predict(customer_train_x)
```

```python
#
silhouette_score(customer_train_x, kmeans.labels_)
```

```python
# finding best value for hyperparameter that best matches apparent clustering
kmeans_per_k = [
    KMeans(n_clusters=k, n_init=100).fit(customer_train_x)
        for k in range(1,10)
]

# first using inertias
inertias = [model.inertia_ for model in kmeans_per_k]

# then using silhouettes
silhouette_scores = [
    silhouette_score(customer_train_x, model.labels_)
        for model in kmeans_per_k[1:]
]
```

```python
print("Inertias:\t", inertias)

plt.plot(range(1, 10), inertias, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Inertia", fontsize=14)
plt.show()

print("Silhouette Scores:\t", silhouette_scores)

plt.plot(range(2, 10), silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette Score", fontsize=14)
plt.show()
```

```python
# now that 3 definitely appears to be the best possible value for 
# the hyperparameter*, retrain the model to cluster with the
# refined k value
#
# *(given the knee at x=3 on the inertial graph
#   and the peak at x=3 on the silhouette graph)
k = 3
kmeans = KMeans(n_clusters=k, n_init=100)
customer_predict_y = kmeans.fit_predict(customer_train_x)

# extracting clusters
customer_clusters = pd.DataFrame()
customer_clusters.index = customer_data_train.index
customer_clusters["cluster"] = customer_predict_y

# saving to CSV
customer_clusters.to_csv("customer_clusters.csv", index=True)

# displaying clusters
pd.set_option("display.max_rows", None, "display.max_columns", None)
for j in range(0, k):
    print(customer_clusters.query("cluster == " + str(j)))
```

### 5. Visualize the results
While clustering appears to be occurring with the training model, it only gives back the customers serial number along with its respective cluster, difficult to visualize.

```python
#
k = 3
reduced_data = PCA(n_components=2).fit_transform(customer_train_x)
kmeans = KMeans(n_clusters=k, n_init=100)
kmeans.fit(reduced_data)

#
h = 0.02

#
x_min, x_max = reduced_data[:,0].min() - 1, reduced_data[:, 0].max() + 1
y_min, y_max = reduced_data[:,1].min() - 1, reduced_data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

#
z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

#
z = z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(
    z, 
    interpolation="nearest",
    extent=(xx.min(), xx.max(), yy.min(), yy.max()),
    cmap=plt.cm.Paired,
    aspect="auto",
    origin="lower",
)

#
plt.plot(reduced_data[:, 0], reduced_data[:, 1], "k", markersize=2)

#
centroids = kmeans.cluster_centers_
plt.scatter(
    centroids[:, 0], 
    centroids[:, 1], 
    marker="x", 
    s=169, 
    linewidths=3, 
    color="w",
    zorder=10,
)

plt.title(
    "K-Means Clustering\n"
    "Centroids marked with white cross"
)

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
